```python
#merge the call_reason and sentiments
call_corr=call_reason.merge(sentiment_statistics,on ='call_id', how='left')
```

```python
call_corr.head()
```

```python
#quantify the percentage difference between the average handling time for the most frequent and least frequent call reasons
call_reason_stats = call_corr.groupby('primary_call_reason').agg(
    count=('primary_call_reason', 'size'),
    avg_AHT=('HT', 'mean')
).reset_index()

# Step 2: Find the most and least frequent call reasons
most_frequent = call_reason_stats.loc[call_reason_stats['count'].idxmax()]
least_frequent = call_reason_stats.loc[call_reason_stats['count'].idxmin()]

# Step 3: Calculate the percentage difference in AHT
most_frequent_AHT = most_frequent['avg_AHT']
least_frequent_AHT = least_frequent['avg_AHT']

percentage_difference = ((most_frequent_AHT - least_frequent_AHT) / most_frequent_AHT) * 100

# Display the results
print(f"Most Frequent Call Reason: {most_frequent['primary_call_reason']} - Average AHT: {most_frequent_AHT}")
print(f"Least Frequent Call Reason: {least_frequent['primary_call_reason']} - Average AHT: {least_frequent_AHT}")
print(f"Percentage Difference in AHT: {percentage_difference:.2f}%")
```

```python
#finding the time at which most calls have to handle group by reasons
call_reason['call_start_datetime'] = pd.to_datetime(call_reason['call_start_datetime'])

# Extract the hour from the call start datetime
call_reason['call_hour'] = call_reason['call_start_datetime'].dt.hour


call_volume = call_reason.groupby(['primary_call_reason', 'call_hour']).size().reset_index(name='call_count')


peak_times = call_volume.loc[call_volume.groupby('primary_call_reason')['call_count'].idxmax()]


print(peak_times)

# Visuallizing
plt.figure(figsize=(12, 6))
sns.lineplot(data=call_volume, x='call_hour', y='call_count', hue='primary_call_reason', marker='o')


plt.title('Peak Call Handling Times by Call Reason')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Calls')
plt.xticks(range(0, 24))  # Set x-ticks to show each hour
plt.legend(title='Call Reason', bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend outside the plot
plt.grid()  # Add grid lines for better readability
plt.tight_layout()  # Adjust layout to fit everything
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
call_corr['customer_tone_encoded'] = encoder.fit_transform(call_corr['customer_tone'])
call_corr['agent_tone_encoded'] = encoder.fit_transform(call_corr['agent_tone'])
call_corr['primary_call_reason_encoded'] = encoder.fit_transform(call_corr['primary_call_reason'])

#Correlation between numeric columns
correlation_matrix = call_corr[['AHT', 'average_sentiment', 'silence_percent_average', 
                         'customer_tone_encoded', 'agent_tone_encoded', 
                         'primary_call_reason_encoded']].corr()


print("Correlation Matrix:")
print(correlation_matrix)

# Visualizng correlation matrix as a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Matrix')
plt.show()

```

```python
 #Grouped by customer tone and call reason, and calculate mean AHT and AST
grouped_data = call_corr.groupby(['customer_tone', 'primary_call_reason']).agg(
    average_AHT=('HT', 'mean'),
    average_AST=('waiting_time', 'mean')
).reset_index()

```

```python
#visualized the data
heatmap_data_aht = grouped_data.pivot("customer_tone", "primary_call_reason", "average_AHT")

# Plot the heatmap for AHT
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap="YlGnBu", annot=True, fmt=".1f")
plt.title('Heatmap of Average Handling Time (AHT) by Customer Tone and Call Reason')
plt.xlabel('Call Reason')
plt.ylabel('Customer Tone')
plt.show()

# Similarly for AST
heatmap_data_ast = grouped_data.pivot("customer_tone", "primary_call_reason", "average_AST")

plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data_ast, cmap="OrRd", annot=True, fmt=".1f")
plt.title('Heatmap of Average Standing Time (AST) by Customer Tone and Call Reason')
plt.xlabel('Call Reason')
plt.ylabel('Customer Tone')
plt.show()

print("Heatmap Data for Average Handling Time (AHT):")
print(heatmap_data_aht)
print("Heatmap Data for Average Standing Time (AST):")
print(heatmap_data_ast)
```
