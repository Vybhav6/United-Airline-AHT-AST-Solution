```python
#converting call_time_data into date_time format 
call_reason['call_start_datetime'] = pd.to_datetime(call_reason['call_start_datetime'], errors='coerce')
call_reason['call_end_datetime'] = pd.to_datetime(call_reason['call_end_datetime'], errors='coerce')
call_reason['primary_call_reason'] = call_reason['primary_call_reason'].str.strip()
call_reason['primary_call_reason'] = (
    call_reason['primary_call_reason']
    .str.replace(r'-', '', regex=True)  # Remove hyphens
    .str.replace(r'\s+', ' ', regex=True)  # Replace multiple spaces with a single space
    .str.strip()  # Remove leading and trailing spaces
)
call_reason['primary_call_reason'] = call_reason['primary_call_reason'].str.replace(
    'Products & Services', 
    'Products and Services', 
    regex=False  # Set regex=False to treat it as a literal string
)
call_reason['primary_call_reason'] = call_reason['primary_call_reason'].str.replace(
    'PostFlight', 
    'Post Flight', 
    regex=False  # Set regex=False to treat it as a literal string
)
```

```python
#quantify the percentage difference between the average handling time for the most frequent and least frequent call reasons
call_reason_stats = call_reason.groupby('primary_call_reason').agg(
    average_handling_time=('HT', 'mean'),
    call_count=('call_id', 'count')
).reset_index()

# Find the most and least frequent call reasons
most_frequent_reason = call_reason_stats.loc[call_reason_stats['call_count'].idxmax()]
least_frequent_reason = call_reason_stats.loc[call_reason_stats['call_count'].idxmin()]

# Calculate the percentage difference in AHT
most_frequent_aht = most_frequent_reason['average_handling_time']
least_frequent_aht = least_frequent_reason['average_handling_time']


if least_frequent_aht > 0:  # To avoid division by zero
    percentage_difference = ((most_frequent_aht - least_frequent_aht) / least_frequent_aht) * 100
else:
    percentage_difference = float('inf')  # Handle the case where least frequent AHT is zero

print(f"Most Frequent Call Reason: {most_frequent_reason['primary_call_reason']}, AHT: {most_frequent_aht:.2f} minutes")
print(f"Least Frequent Call Reason: {least_frequent_reason['primary_call_reason']}, AHT: {least_frequent_aht:.2f} minutes")
print(f"Percentage Difference in AHT: {percentage_difference:.2f}%")
```

```python
#finding the time at which most calls have to handle group by reasons
call_reason['call_start_datetime'] = pd.to_datetime(call_reason['call_start_datetime'])

# Extract the hour from the call start datetime
call_reason['call_hour'] = call_reason['call_start_datetime'].dt.hour


call_volume = call_reason.groupby(['primary_call_reason', 'call_hour']).size().reset_index(name='call_count')


peak_times = call_volume.loc[call_volume.groupby('primary_call_reason')['call_count'].idxmax()]


print(peak_times)

# Visuallizing
plt.figure(figsize=(12, 6))
sns.lineplot(data=call_volume, x='call_hour', y='call_count', hue='primary_call_reason', marker='o')


plt.title('Peak Call Handling Times by Call Reason')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Calls')
plt.xticks(range(0, 24))  # Set x-ticks to show each hour
plt.legend(title='Call Reason', bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend outside the plot
plt.grid()  # Add grid lines for better readability
plt.tight_layout()  # Adjust layout to fit everything
plt.show()
```
